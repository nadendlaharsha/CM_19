```python
from rouge_score import rouge_scorer
import pandas as pd

def calculate_rouge(reference, candidate):
    """Calculates ROUGE-1, ROUGE-2, and ROUGE-L F1 scores."""
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, candidate)
    
    # Returning F1-Score scaled to 0-100% (Standard for Research Papers)
    return {
        'ROUGE-1 Precision': round(scores['rouge1'].precision * 100, 2),
        'ROUGE-1 Recall': round(scores['rouge1'].recall * 100, 2),
        'ROUGE-1 F1': round(scores['rouge1'].fmeasure * 100, 2),
        'ROUGE-2 F1': round(scores['rouge2'].fmeasure * 100, 2),
        'ROUGE-L F1': round(scores['rougeL'].fmeasure * 100, 2)
    }
```

```python
# 2. Define Summaries

# (A) Human-Written Ground Truth Summary (Reference)
reference_summary = """
 The speaker says that managing the audio-visual information in the videos may be difficult, that is why the presence of a transcript in written form is valuable for summarizing and remembering the information.
"""

# (B) Summary from Previous Baseline Papers
previous_paper_summary = """
 the speaker explains that verbal and visual information in videos can be difficult to manage, which is why having written transcripts is so valuable for summarization and retention.
"""

# (C) Summary generated by YOUR Project (Proposed Method)
my_project_summary ="""In this video, Demitri shares three simple hacks for obtaining and summarizing YouTube video transcripts. He addresses the difficulty of processing purely verbal and visual information, The speaker says that managing the audio-visual information in the videos may be difficult, that is why the presence of a transcript in written form is valuable for summarizing and remembering the information."""
```

```python
# 3. Calculate Scores and Generate Table
baseline_scores = calculate_rouge(reference_summary, previous_paper_summary)
my_project_scores = calculate_rouge(reference_summary, my_project_summary)

compare_data = [
    {
        "Model / Approach": "Previous Paper (Baseline)", 
        "ROUGE-1 (F1)": baseline_scores['ROUGE-1 F1'],
        "ROUGE-2 (F1)": baseline_scores['ROUGE-2 F1'],
        "ROUGE-L (F1)": baseline_scores['ROUGE-L F1']
    },
    {
        "Model / Approach": "Proposed Method (Our Project)", 
        "ROUGE-1 (F1)": my_project_scores['ROUGE-1 F1'],
        "ROUGE-2 (F1)": my_project_scores['ROUGE-2 F1'],
        "ROUGE-L (F1)": my_project_scores['ROUGE-L F1']
    }
]

results_df = pd.DataFrame(compare_data)

print("================ PERFORMANCE COMPARISON TABLE =================")
print(results_df.to_string(index=False))
print("===============================================================\n")
```
